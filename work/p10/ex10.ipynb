## Probability distribution functions

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

Considering a probability distribution $p\left(x\right) = bx$ in the domain $\left[0,a\right\rangle$.

We consider the integral of our probability distribution over our domain. This gives us $\int_0^abx\mathrm{d}x=a^2\frac{b}{2}$, which is equal to 1, since it is supposed to be normalized. Solving for b results in: $b= \frac{2}{a^2}$.

```python
a = 0.5

def p(x, a = a):
    b = 2/(a**2)
    p = b*x
    return p 

r = np.random.uniform(0,1,10000)
s = np.random.uniform(0,1,10000)
x = a*r
y = p(a)*s

val = []
ind = []
for i in range(0, len(r)):
    if y[i]<=p(x)[i]:
        val.append(x[i])
        ind.append(True)
    else: ind.append(False)
    
val[0:10]
```

```python
# Plotting the histogram
num_bins = 30  # Number of bins for the histogram
plt.hist(val, bins=num_bins, density=True, label='Histogram')

x = np.linspace(0, a, 1000)
b = 2 / (a**2)  
p_x = b * x  
plt.plot(x, p_x, 'r', label='Theoretical Distribution')

plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
# TODO: plot for different sample sizes
```

## Determine the number $\pi$ with random numbers RN

```python
def f(x):
    f = np.sqrt(1-x**2)
    return(f)

n = np.logspace(1, 5, num= 8)
x = np.random.uniform(0,1,int(max(n)))
y = np.random.uniform(0,1,int(max(n)))
```

```python

i = y<=f(x)
p = []
for j in range(0, len(n)):
    N = int(n[j])
    P = sum(i[0:N])
    p.append(P)

PI = p/n*4
```

```python
f, ax = plt.subplots(4,2,figsize=(8,8))
for j in range(0,len(n)):
    N = int(n[j])
    row = j // 2
    col = j % 2
    ax[row, col].scatter(x[:N], y[:N], c=i[:N])
print(PI)
```
