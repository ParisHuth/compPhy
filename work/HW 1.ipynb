```python
import numpy as np
import matplotlib.pyplot as plt

# Numerical Integration

def inte(x,a,n):
    return x**n/(x+a)
    
a = 5
x = np.linspace(0,1,100)
n = np.array([1,5,10,20,30,50])

y = np.array([inte(x,a,i) for i in n])

for i in range(len(n)):
    plt.plot(x,y[i], label = r'f({}, {}, x)'.format(i,a))

plt.legend()
plt.grid()
plt.show()  
```

```python
def iteration(a,n0,n1,y0):
    y_n = None
    if n0 < n1:
        un = None
        y_n = np.array([y0])
        for n in range(n0,n1):
            un = 1/(n+1) - a*y_n[-1]
            y_n = np.append(y_n, un)
    elif n0 > n1:
        un = None
        y_n = np.array([y0])
        for n in range(n1,n0):
            un = 1/a*(1/n - y_n[-1])
            y_n = np.append(y_n, un)
    elif n0 ==n1: 
        y_n = np.array([y0])
        
    return y_n
```

```python
#to study the behavior of the series, the whole iteration was computated, instead of 
#just the last element
a = iteration(5,0,30,np.log((1+5)/5))
for i in range(0,len(a)):
    print('y_{} = {}'.format(0+ i,a[i]))
```

```python
b = iteration(5,50,30,np.log((1+5)/5))
for i in range(0,len(b)):
    print('y_{} = {}'.format(50 - i,b[i]))
```
