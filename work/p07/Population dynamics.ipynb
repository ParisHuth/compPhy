# Population dynamics
### _a)_
The normalized equation reads as:
$$\frac{d n}{d t} = \alpha n(1-n\beta) - \frac{n^2}{1+n^2} $$

The fix points are given by following expression:
$$n\left[ \alpha (1-n \beta) -\frac{n}{1+n^2}\right] = 0$$

The first factorial give the trivial fixpoint:
-$$n_0 = 0$$
and the rest of the equation of the equation can be reduced to:
$$1+n^2-n\left( \frac{1}{\alpha} +\beta\right) -n^3 \beta = 0$$

```python
def roots_f(a):
    
    '''
    This function calculates the roots of the cubic parameter
    Input:
    a - float : paramter alpha on the polynomial
    
    Output:
    roots - n-array containing the 3 remainting roots
    '''
    
    coeff = [-1/8,1,-(1/a+1/8),1]
    roots = np.roots(coeff)
    for i in range(0,len(roots)):
        print("n_{} = {}".format(i+1,roots[i]))
    return roots
```

### _b)_
From part be, one can notice that for $\alpha = 0.5$ the four roots (uncluding trivial $n_0$) are real. 

```python
#plotting F(n) for differnt alpha values

def f(n,alpha):
    return 1-n*(1/alpha +1/8)+n**2 -n**3*1/8
yero = np.zeros(100)


pl.figure(figsize = (9,10))
for a in np.arange(0.2,0.8,0.05):
    
    n = np.linspace(0,6,100)
    plt.plot(n,f(n,a), label= r" $\alpha$ = {}".format(a))
    plt.legend()
    plt.plot(n,yero)
    plt.ylim(-1.5,1.5)
```

The observation made on _a)_ about the real roots gets corroborated by the upper plot. To check the switching point, we repeat the plot near 0.5

```python
#plotting F(n) for differnt 0,4< alpha <0,5 values

def f(n,alpha):
    return 1-n*(1/alpha +1/8)+n**2 -n**3*1/8
yero = np.zeros(100)

pl.figure(figsize = (9,10))
for a in np.arange(0.468,.58,0.01):
    
    n = np.linspace(0,6,100)
    pl.plot(n,f(n,a), label= r" $\alpha$ = {}".format(a))
    pl.legend(loc="best")
    pl.plot(n,yero, color = "red")
    pl.ylim(-1.5,1.5)
```

For values $0,685 < \alpha < 0,58$ $F(n)$ has three real fixpoints

```python
def  df(n,a):
    return a - 2*n*a*1/8 - 2*n/((1+n**2)**2)

def roots_f(a):
    
    coeff = [-1/8,1,-(1/a+1/8),1]
    roots = np.roots(coeff)

    return roots

def check(a):
    print("for alpha = {}".format(a))
    roots = roots_f(a)
    for i in range(0,len(roots)):
        if roots[i].imag == 0:
            df_roots = df(roots[i].real,a)
            if df_roots < 0:
                print("n_{} = {} is stable".format(i+1,roots[i]))
            elif df_roots > 0:
                print("n_{} = {} is unstable".format(i+1,roots[i]))
            else:
                print(" the stability of n_{} = {} couldn't be determinated")
        else:
            pass
    print()
    
```
